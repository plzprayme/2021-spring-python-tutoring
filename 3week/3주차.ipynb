```python
x, y = 0, 0
while y < 5:

  while x < 5:
    print('*', end="")

    if x == y:
      x = 0
      break

    x += 1

  print()
  y += 1
```

```python
# 프로그래머스 콜라츠 추측
# 1. while문 활용한 풀이

num = 6      # 정답: 8
# num = 16     # 정답: 4
# num = 626331 # -1

answer = 0 
while True: 
    if num == 1: 
        return answer
    elif answer == 500: #500회 일때 -1 리턴
        return -1

    if num % 2 == 0:  
        num = num / 2
    elif num % 2 == 1: 
        num = num * 3 + 1

    answer += 1
return answer

#2. for문 활용한 풀이
for i in range(500):
  if num == 1:
      return i

  if num % 2 == 0: # 1-1. 입력된 수가 짝수라면 2로 나눕니다.
      num //= 2
  else:       # 1-2. 입력된 수가 홀수라면 3을 곱하고 1을 더합니다.
      num = num * 3 + 1
  return -1
```

```python
# 백준 온라인저지지 10871번 문제
# 1. list comprehension을 이용한 풀이
N, X = map(int, input().split())
result = [i for i in map(int, input().split()) if i < X]
[print(i, end=' ') for i in result]
print(result[-1], end=' ')

# 2. filter를 활용한 풀이
N, X = map(int, input().split())
A = map(int, input().split())
[print(i, end=' ') for i in filter(lambda i: i < X, A)]
```
